```python
import pandas as pd
df_fear_greed = pd.read_csv('/content/fear_greed_index.csv')
df_trader = pd.read_csv('/content/historical_data.csv')
```

```python
print(df_fear_greed.info())
print(df_fear_greed.head())
```

```python
print(df_trader.info())
print(df_trader.head())
```

```python
df_fear_greed['date'] = pd.to_datetime(df_fear_greed['date'])
df_trader['Timestamp_IST'] = pd.to_datetime(df_trader['Timestamp IST'],format='%d-%m-%Y %H:%M')
df_trader.drop(columns=['Timestamp IST'],inplace=True)
```

```python
df_trader['date'] = df_trader['Timestamp_IST'].dt.date
df_trader['date'] = pd.to_datetime(df_trader['date'])
trader_subset = df_trader[['Account', 'Coin', 'date', 'Closed PnL', 'Size USD', 'Side', 'Direction', 'Fee', 'Execution Price', 'Timestamp_IST']]
```

```python
df_merged = trader_subset.merge(df_fear_greed[['date', 'value', 'classification']],
                               on='date', how='left')

print(f"Merged data shape: {df_merged.shape}")
print(df_merged[['date', 'classification', 'Closed PnL', 'Size USD']].head())
```

```python
print(df_merged['classification'].value_counts(dropna=False))
```

```python

import plotly.express as px

# Group and aggregate
summary = df_merged.groupby('classification').agg({
    'Closed PnL': ['mean', 'median', 'std'],
    'Size USD': ['mean', 'median', 'sum'],
}).reset_index()
# Flatten MultiIndex columns
summary.columns = ['classification', 'ClosedPnL_mean', 'ClosedPnL_median', 'ClosedPnL_std',
                   'SizeUSD_mean', 'SizeUSD_median', 'SizeUSD_sum']

print(summary)
```

```python
fig = px.box(df_merged, x='classification', y='Closed PnL',
             title='Trade Profit Distribution by Market Sentiment',
             points='all', color='classification')
fig.show()
```

now Visualizing trading volume by sentiment


```python
fig = px.bar(summary, x='classification', y='SizeUSD_sum',
             title='Total Trading Volume by Market Sentiment',
             labels={'SizeUSD_sum': 'Total Size USD'})
fig.show()
```

now Outliers or Anomalies

```python
top_pnl = df_merged.nlargest(10, 'Closed PnL')
print(top_pnl[['date', 'classification', 'Closed PnL', 'Size USD']])

bottom_pnl = df_merged.nsmallest(10, 'Closed PnL')
print(bottom_pnl[['date', 'classification', 'Closed PnL', 'Size USD']])
```

```python
# Example: Avg Closed PnL by Account and Sentiment
account_summary = df_merged.groupby(['Account', 'classification'])['Closed PnL'].mean().reset_index()

# Example: Rolling mean of Closed PnL over time
df_merged['date'] = pd.to_datetime(df_merged['date'])
daily_pnl = df_merged.groupby(['date', 'classification'])['Closed PnL'].mean().reset_index()
daily_pnl['rolling_mean'] = daily_pnl.groupby('classification')['Closed PnL'].transform(lambda x: x.rolling(7).mean())

import plotly.express as px
fig = px.line(daily_pnl, x='date', y='rolling_mean', color='classification',
              title='7-Day Rolling Average Closed PnL by Sentiment')
fig.show()
```

```python
# Average Closed PnL per account and sentiment
account_summary = (
    df_merged
    .groupby(['Account', 'classification'])['Closed PnL']
    .mean()
    .reset_index()
)

# Top 10 performing traders in 'Greed'
top_greed = account_summary[account_summary['classification'] == 'Greed'].nlargest(10, 'Closed PnL')
print("Top traders in Greed regimes:")
print(top_greed)
```

```python
# Average Closed PnL per coin and sentiment
coin_summary = (
    df_merged
    .groupby(['Coin', 'classification'])['Closed PnL']
    .mean()
    .reset_index()
)

# Plot with Plotly (for coins with most trades)

top_coins = df_merged['Coin'].value_counts().head(5).index.tolist()
fig = px.bar(
    coin_summary[coin_summary['Coin'].isin(top_coins)],
    x='Coin', y='Closed PnL', color='classification',
    barmode='group',
    title='Average Closed PnL by Coin and Sentiment'
)
fig.show()
```

```python
direction_summary = (
    df_merged
    .groupby(['Direction', 'classification'])['Closed PnL']
    .mean()
    .reset_index()
)

fig = px.bar(
    direction_summary,
    x='Direction', y='Closed PnL', color='classification',
    barmode='group',
    title='Avg Closed PnL by Direction and Sentiment'
)
fig.show()
```

```python
from scipy.stats import ttest_ind

greed_pnl = df_merged[df_merged['classification'] == 'Greed']['Closed PnL']
fear_pnl = df_merged[df_merged['classification'] == 'Fear']['Closed PnL']
stat, pval = ttest_ind(greed_pnl, fear_pnl, nan_policy='omit', equal_var=False)
print(f"T-test Greed vs Fear: statistic={stat:.2f}, p-value={pval:.3g}")
```
